```python
import os
import time
import glob
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys

# =============================
# CONFIG
# =============================
EMAIL = "trinm1010@gmail.com"
PASSWORD = "Tri1010"

DOWNLOAD_DIR = os.path.abspath("downloads")
OUTPUT_FILE = "macro_combined.csv"

MACRO_URL = "https://finance.vietstock.vn/du-lieu-vi-mo"

# =============================
# SETUP SELENIUM
# =============================
options = Options()
options.add_argument("--start-maximized")
options.add_experimental_option("prefs", {
    "download.default_directory": DOWNLOAD_DIR,
    "download.prompt_for_download": False,
    "download.directory_upgrade": True,
    "safebrowsing.enabled": True
})
driver = webdriver.Chrome(options=options)

# =============================
# LOGIN
# =============================
driver.get("https://finance.vietstock.vn/Login")

time.sleep(2)
driver.find_element(By.ID, "txtEmail").send_keys(EMAIL)
driver.find_element(By.ID, "txtPassword").send_keys(PASSWORD)
driver.find_element(By.ID, "btnLogin").click()

time.sleep(3)  # ch·ªù login xong

# =============================
# V√ÄO TRANG D·ªÆ LI·ªÜU Vƒ® M√î
# =============================
driver.get(MACRO_URL)
time.sleep(3)

# =============================
# DANH S√ÅCH C√ÅC CH·ªà S·ªê C·∫¶N DOWNLOAD
# =============================
# ID c·ªßa t·ª´ng ch·ªâ s·ªë tr√™n sidebar ‚Äì b·∫°n c√≥ th·ªÉ th√™m b·ªõt t√πy √Ω
macro_items = [
    "GDP",
    "CPI",
    "Trade",
    "FDI",
    "Retail",
    "IndustrialProduction"
]

# =============================
# H√ÄM CLICK V√Ä DOWNLOAD
# =============================
def download_macro(name):
    print(f"‚ñ∂ Download: {name}")

    # Click menu tr√°i (s·ª≠ d·ª•ng XPATH linh ho·∫°t)
    menu_xpath = f"//a[contains(text(), '{name}')]"
    driver.find_element(By.XPATH, menu_xpath).click()
    time.sleep(2)

    # Click n√∫t export Excel
    export_btn = driver.find_element(By.XPATH, "//button[contains(text(),'Xu·∫•t d·ªØ li·ªáu')]")
    export_btn.click()

    print("   ‚Üí ƒê√£ b·∫•m t·∫£i Excel...")
    time.sleep(4)  # ch·ªù file t·∫£i xong


# =============================
# TI·∫æN H√ÄNH T·∫¢I T·∫§T C·∫¢
# =============================
for item in macro_items:
    download_macro(item)

driver.quit()

# =============================
# G·ªòP T·∫§T C·∫¢ FILE EXCEL
# =============================
files = glob.glob(os.path.join(DOWNLOAD_DIR, "*.xlsx"))

df_all = []
for f in files:
    try:
        df = pd.read_excel(f)
        df["source_file"] = os.path.basename(f)
        df_all.append(df)
    except:
        print("L·ªói khi ƒë·ªçc:", f)

# Combine
df_final = pd.concat(df_all, ignore_index=True)
df_final.to_csv(OUTPUT_FILE, index=False, encoding="utf-8-sig")

print("üéâ DONE! D·ªØ li·ªáu t·ªïng h·ª£p l∆∞u t·∫°i:", OUTPUT_FILE)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

EMAIL = "trinm1010@gmail.com"
PASSWORD = "Tri1010"

options = Options()
options.add_argument("--start-maximized")
driver = webdriver.Chrome(options=options)

wait = WebDriverWait(driver, 20)

driver.get("https://finance.vietstock.vn")

# ================================
# 1) CLICK N√öT ƒêƒÇNG NH·∫¨P
# ================================
login_btn = wait.until(EC.element_to_be_clickable((
    By.XPATH, "//a[contains(@class,'btnlogin-link')]"
)))
login_btn.click()

# *** IMPORTANT ***
# Kh√¥ng click g√¨ n·ªØa, ch·ªù popup m·ªü ra

# ================================
# 2) CH·ªú POPUP HI·ªÜN (modal in)
# ================================
wait.until(EC.visibility_of_element_located((
    By.XPATH, "//div[@id='login-form' and contains(@class,'in')]"
)))

# ================================
# 3) ƒêI·ªÄN EMAIL
# ================================
email_field = wait.until(EC.presence_of_element_located((
    By.XPATH, "//div[@id='login-form']//input[@type='email']"
)))
email_field.clear()
email_field.send_keys(EMAIL)

# ================================
# 4) ƒêI·ªÄN PASSWORD
# ================================
password_field = wait.until(EC.presence_of_element_located((
    By.XPATH, "//div[@id='login-form']//input[@type='password']"
)))
password_field.clear()
password_field.send_keys(PASSWORD)

# ================================
# 5) CLICK LOGIN
# ================================
submit_btn = wait.until(EC.element_to_be_clickable((
    By.XPATH, "//div[@id='login-form']//button[contains(text(),'ƒêƒÉng nh·∫≠p')]"
)))
submit_btn.click()

time.sleep(5)

print("üéâ ƒêƒÇNG NH·∫¨P TH√ÄNH C√îNG!")
```
