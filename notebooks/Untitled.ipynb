```python
import requests
import certifi
import pandas as pd

def wb_get(indicator):
    url = f"https://api.worldbank.org/v2/country/VN/indicator/{indicator}?format=json&per_page=2000"
    response = requests.get(url, verify=certifi.where())
    data = response.json()
    df = pd.DataFrame(data[1])
    df['value'] = df['value'].astype(float)
    return df[['date','value']]

gdp = wb_get("NY.GDP.MKTP.CD")
print(gdp.head())
```
