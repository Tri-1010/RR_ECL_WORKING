```python


# Import
from vnstock_data import Macro
marco = Marco(source = 'mbk')
import pandas as pd

# =============================
# 1. L·∫°m ph√°t (CPI)
# =============================
cpi = macro.cpi()
print("CPI:")
print(cpi.head())

# =============================
# 2. GDP theo qu√Ω
# =============================
gdp_quarter = macro.gdp_quarter()
print("GDP QUARTER:")
print(gdp_quarter.head())

# =============================
# 3. GDP theo nƒÉm
# =============================
gdp_year = macro.gdp_year()
print("GDP YEAR:")
print(gdp_year.head())

# =============================
# 4. L√£i su·∫•t ƒëi·ªÅu h√†nh (SBV)
# =============================
interest_rate = macro.interest_rates()
print("Interest Rates:")
print(interest_rate.head())

# =============================
# 5. T·ª∑ gi√° USD/VND
# =============================
usd_vnd = macro.exchange_rates()
print("USD/VND:")
print(usd_vnd.head())

# =============================
# 6. D·ª± tr·ªØ ngo·∫°i h·ªëi
# =============================
fx_reserves = macro.fx_reserves()
print("FX reserves:")
print(fx_reserves.head())

# =============================
# 7. T·ªïng ph∆∞∆°ng ti·ªán thanh to√°n M2
# =============================
money_supply = macro.money_supply()
print("M2:")
print(money_supply.head())

# =============================
# 8. T√≠n d·ª•ng theo ng√†nh
# =============================
credit_growth = macro.credit_growth()
print("Credit growth:")
print(credit_growth.head())

# =============================
# 9. D√¢n s·ªë Vi·ªát Nam
# =============================
population = macro.population()
print("Population:")
print(population.head())

# =============================
# 10. Xu·∫•t nh·∫≠p kh·∫©u Vi·ªát Nam (Trade)
# =============================
trade = macro.trade_balance()
print("Trade:")
print(trade.head())

# =============================
# 11. PMI s·∫£n xu·∫•t (Vietnam PMI)
# =============================
pmi = macro.pmi()
print("PMI:")
print(pmi.head())

# =============================
# Xu·∫•t to√†n b·ªô d·ªØ li·ªáu ra file Excel
# =============================
with pd.ExcelWriter("vietnam_macro_vnstock.xlsx") as writer:
    cpi.to_excel(writer, sheet_name="CPI", index=False)
    gdp_quarter.to_excel(writer, sheet_name="GDP_Quarter", index=False)
    gdp_year.to_excel(writer, sheet_name="GDP_Year", index=False)
    interest_rate.to_excel(writer, sheet_name="Interest_Rates", index=False)
    usd_vnd.to_excel(writer, sheet_name="USDVND", index=False)
    fx_reserves.to_excel(writer, sheet_name="FX_reserves", index=False)
    money_supply.to_excel(writer, sheet_name="M2", index=False)
    credit_growth.to_excel(writer, sheet_name="Credit_Growth", index=False)
    population.to_excel(writer, sheet_name="Population", index=False)
    trade.to_excel(writer, sheet_name="Trade", index=False)
    pmi.to_excel(writer, sheet_name="PMI", index=False)

print(">>> Export completed: vietnam_macro_vnstock.xlsx")
```

```python
import pandas as pd
# 1. Th∆∞ vi·ªán l·∫•y d·ªØ li·ªáu Vƒ© m√¥ t·ª´ World Bank
from pandas_datareader import wb
# 2. Th∆∞ vi·ªán l·∫•y d·ªØ li·ªáu Ch·ª©ng kho√°n Vi·ªát Nam
from vnstock import stock_historical_data

print("--- B·∫ÆT ƒê·∫¶U T·∫¢I D·ªÆ LI·ªÜU ---\n")

# -----------------------------------------------------------
# PH·∫¶N 1: L·∫§Y D·ªÆ LI·ªÜU Vƒ® M√î VI·ªÜT NAM (Ngu·ªìn: World Bank)
# -----------------------------------------------------------
# M√£ ch·ªâ s·ªë World Bank:
# - NY.GDP.MKTP.KD.ZG: TƒÉng tr∆∞·ªüng GDP (%)
# - FP.CPI.TOTL.ZG: L·∫°m ph√°t CPI (%)
# - FR.INR.LEND: L√£i su·∫•t cho vay (%)

try:
    print("1. ƒêang t·∫£i d·ªØ li·ªáu Vƒ© m√¥ t·ª´ World Bank...")
    # T·∫£i d·ªØ li·ªáu t·ª´ nƒÉm 2015 ƒë·∫øn 2023 c·ªßa Vi·ªát Nam (M√£ qu·ªëc gia: VN)
    df_macro = wb.download(indicator=['NY.GDP.MKTP.KD.ZG', 'FP.CPI.TOTL.ZG', 'FR.INR.LEND'], 
                           country=['VN'], 
                           start=2015, 
                           end=2024)
    
    # ƒê·ªïi t√™n c·ªôt cho d·ªÖ ƒë·ªçc
    df_macro = df_macro.rename(columns={
        'NY.GDP.MKTP.KD.ZG': 'Tang_truong_GDP (%)', 
        'FP.CPI.TOTL.ZG': 'Lam_phat_CPI (%)',
        'FR.INR.LEND': 'Lai_suat_cho_vay (%)'
    })
    print(df_macro)
    print("-" * 30)
except Exception as e:
    print(f"L·ªói t·∫£i d·ªØ li·ªáu World Bank: {e}")

# -----------------------------------------------------------
# PH·∫¶N 2: L·∫§Y D·ªÆ LI·ªÜU CH·ª®NG KHO√ÅN (Ngu·ªìn: Vnstock)
# -----------------------------------------------------------

try:
    print("\n2. ƒêang t·∫£i d·ªØ li·ªáu Ch·ª©ng kho√°n t·ª´ Vnstock...")
    # L·∫•y l·ªãch s·ª≠ gi√° VCB t·ª´ ƒë·∫ßu nƒÉm 2024 ƒë·∫øn nay
    df_stock = stock_historical_data(symbol='VCB', 
                                   start_date='2024-01-01', 
                                   end_date='2024-11-27',
                                   resolution='1D', 
                                   type='stock')
    
    if df_stock is not None and not df_stock.empty:
        print(df_stock.head())
    else:
        print("Kh√¥ng t·∫£i ƒë∆∞·ª£c d·ªØ li·ªáu ch·ª©ng kho√°n.")
        
except NameError:
    print("L·ªói: H√†m 'stock_historical_data' kh√¥ng t√¨m th·∫•y. H√£y ki·ªÉm tra l·∫°i version vnstock.")
except Exception as e:
    print(f"L·ªói kh√°c: {e}")

print("\n--- HO√ÄN T·∫§T ---")
```

```python
import requests
import pandas as pd

def gso_cpi():
    url = "https://api.gso.gov.vn/api/so-lieu/142"   # ch·ªâ s·ªë CPI theo th√°ng
    r = requests.get(url)
    data = r.json()

    rows = []
    for item in data['data']:
        rows.append({
            "Year": item['nam'],
            "Month": item['thang'],
            "CPI": item['giatri']
        })
    return pd.DataFrame(rows)

df_cpi = gso_cpi()
df_cpi.head()
```
