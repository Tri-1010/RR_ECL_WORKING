```python
from vnstock import *
import pandas as pd

# C·∫•u h√¨nh hi·ªÉn th·ªã pandas cho d·ªÖ ƒë·ªçc
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print("--- ƒêANG T·∫¢I D·ªÆ LI·ªÜU Vƒ® M√î T·ª™ VNSTOCK ---\n")

# -----------------------------------------------------------
# NH√ìM 1: D·ªÆ LI·ªÜU THEO TH√ÅNG (Monthly)
# Quan tr·ªçng: CPI (L·∫°m ph√°t) & IIP (S·∫£n xu·∫•t)
# -----------------------------------------------------------

# 1. Ch·ªâ s·ªë gi√° ti√™u d√πng (CPI)
# D·ªØ li·ªáu tr·∫£ v·ªÅ s·∫Ω c√≥ c√°c nh√≥m ng√†nh: H√†ng ƒÉn, Gi√°o d·ª•c, Y t·∫ø, Giao th√¥ng...
try:
    df_cpi = macro.get_cpi_data(start_date='2020-01-01', end_date='2024-11-27')
    print("1. D·ªØ li·ªáu CPI (Theo th√°ng):")
    print(df_cpi.tail(5)) # Xem 5 d√≤ng cu·ªëi
    print("-" * 30)
except Exception as e:
    print(f"L·ªói l·∫•y CPI: {e}")

# 2. Ch·ªâ s·ªë s·∫£n xu·∫•t c√¥ng nghi·ªáp (IIP)
# Ph·∫£n √°nh s·ª©c kh·ªèe c·ªßa n·ªÅn kinh t·∫ø s·∫£n xu·∫•t
try:
    df_iip = macro.get_industrial_index(start_date='2020-01-01', end_date='2024-11-27')
    print("2. D·ªØ li·ªáu IIP (Theo th√°ng):")
    print(df_iip.tail(5))
    print("-" * 30)
except Exception as e:
    print(f"L·ªói l·∫•y IIP: {e}")

# -----------------------------------------------------------
# NH√ìM 2: D·ªÆ LI·ªÜU THEO QU√ù (Quarterly)
# Quan tr·ªçng: GDP
# -----------------------------------------------------------

# 3. TƒÉng tr∆∞·ªüng GDP
try:
    df_gdp = macro.get_gdp_data(start_date='2020-01-01', end_date='2024-11-27')
    print("3. D·ªØ li·ªáu GDP (Theo qu√Ω):")
    # L∆∞u √Ω: C·ªôt th·ªùi gian th∆∞·ªùng s·∫Ω l√† d·∫°ng "Q1-2023", "Q2-2023"
    print(df_gdp.tail(5))
    print("-" * 30)
except Exception as e:
    print(f"L·ªói l·∫•y GDP: {e}")

# -----------------------------------------------------------
# NH√ìM 3: TI·ªÄN T·ªÜ & T·ª∂ GI√Å (C·ª±c quan tr·ªçng cho T√†i ch√≠nh Ti√™u d√πng)
# -----------------------------------------------------------

# 4. T·ª∑ gi√° trung t√¢m (D·ªØ li·ªáu h√†ng ng√†y - Daily)
try:
    df_fx = stock_historical_data(symbol='USD/VND', 
                                start_date='2024-01-01', 
                                end_date='2024-11-27', 
                                resolution='1D', 
                                type='index') # index cho t·ª∑ gi√°
    print("4. T·ª∑ gi√° USD/VND (H√†ng ng√†y):")
    print(df_fx.tail(5))
    print("-" * 30)
except Exception as e:
    print(f"L·ªói l·∫•y t·ª∑ gi√°: {e}")

# -----------------------------------------------------------
# TIP CHO NG∆Ø·ªúI L√ÄM DATA:
# D·ªØ li·ªáu vƒ© m√¥ th∆∞·ªùng c√≥ c·ªôt ng√†y th√°ng l·ªôn x·ªôn (l√∫c th√¨ text, l√∫c th√¨ date).
# H√£y d√πng l·ªánh n√†y ƒë·ªÉ chu·∫©n h√≥a v·ªÅ datetime ƒë·ªÉ d·ªÖ plot bi·ªÉu ƒë·ªì.
# -----------------------------------------------------------

if 'df_cpi' in locals() and not df_cpi.empty:
    # V√≠ d·ª• v·ªõi CPI, c·ªôt ng√†y th∆∞·ªùng t√™n l√† 'period' ho·∫∑c 'date'
    # B·∫°n c·∫ßn ki·ªÉm tra t√™n c·ªôt th·ª±c t·∫ø khi ch·∫°y l·ªánh print(df_cpi.columns)
    # Gi·∫£ s·ª≠ c·ªôt ch·ª©a ng√†y l√† 'period' v√† c√≥ d·∫°ng '2024-10'
    print("\n--- TIP: Ki·ªÉm tra t√™n c·ªôt ƒë·ªÉ convert ---")
    print(f"C√°c c·ªôt trong b·∫£ng CPI: {df_cpi.columns.tolist()}")
```

```python
import pandas as pd
from pandas_datareader import wb
# Import c·ª• th·ªÉ h√†m t·ª∑ gi√° t·ª´ vnstock ƒë·ªÉ tr√°nh l·ªói
from vnstock import stock_historical_data

print("--- ƒêANG T·∫¢I D·ªÆ LI·ªÜU ---")

# ==========================================
# 1. D·ªÆ LI·ªÜU Vƒ® M√î (T·ª™ WORLD BANK)
# ==========================================
try:
    print("\n1. ƒêang l·∫•y d·ªØ li·ªáu Vƒ© m√¥ (GDP, CPI, L√£i su·∫•t)...")
    # L·∫•y d·ªØ li·ªáu t·ª´ 2018 ƒë·∫øn 2023
    df_macro = wb.download(indicator=['NY.GDP.MKTP.KD.ZG', 'FP.CPI.TOTL.ZG', 'FR.INR.LEND'], 
                           country=['VN'], 
                           start=2018, 
                           end=2023)
    
    # ƒê·ªïi t√™n c·ªôt ti·∫øng Vi·ªát
    df_macro = df_macro.rename(columns={
        'NY.GDP.MKTP.KD.ZG': 'Tang_truong_GDP (%)', 
        'FP.CPI.TOTL.ZG': 'Lam_phat_CPI (%)',
        'FR.INR.LEND': 'Lai_suat_cho_vay (%)'
    })
    print(df_macro)
except Exception as e:
    print(f"L·ªói World Bank: {e}")

# ==========================================
# 2. D·ªÆ LI·ªÜU T·ª∂ GI√Å (T·ª™ VNSTOCK)
# ==========================================
try:
    print("\n2. ƒêang l·∫•y d·ªØ li·ªáu T·ª∑ gi√° USD/VND...")
    # L·∫•y t·ª∑ gi√° t·ª´ ƒë·∫ßu nƒÉm 2024
    df_fx = stock_historical_data(symbol='USD/VND', 
                                start_date='2024-01-01', 
                                end_date='2024-11-27', 
                                resolution='1D', 
                                type='index')
    if df_fx is not None and not df_fx.empty:
        print(df_fx.tail())
    else:
        print("Kh√¥ng c√≥ d·ªØ li·ªáu t·ª∑ gi√° tr·∫£ v·ªÅ.")
except Exception as e:
    print(f"L·ªói Vnstock: {e}")

print("\n--- HO√ÄN T·∫§T ---")
```

```python
import pandas as pd
from pandas_datareader import wb
# L√∫c n√†y import s·∫Ω kh√¥ng c√≤n b·ªã l·ªói n·ªØa
from vnstock import stock_historical_data

# C·∫•u h√¨nh hi·ªÉn th·ªã r·ªông ƒë·ªÉ d·ªÖ nh√¨n
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print("--- B·∫ÆT ƒê·∫¶U T·∫¢I D·ªÆ LI·ªÜU ---\n")

# =======================================================
# PH·∫¶N 1: D·ªÆ LI·ªÜU Vƒ® M√î (GDP, CPI, L√ÉI SU·∫§T) - Ngu·ªìn: World Bank
# =======================================================
try:
    print("1. ƒêang t·∫£i d·ªØ li·ªáu Vƒ© m√¥ t·ª´ World Bank...")
    # L·∫•y d·ªØ li·ªáu t·ª´ 2018 ƒë·∫øn 2023
    # M√£: GDP tƒÉng tr∆∞·ªüng, CPI l·∫°m ph√°t, L√£i su·∫•t cho vay
    df_macro = wb.download(indicator=['NY.GDP.MKTP.KD.ZG', 'FP.CPI.TOTL.ZG', 'FR.INR.LEND'], 
                           country=['VN'], 
                           start=2018, 
                           end=2023)
    
    # ƒê·ªïi t√™n c·ªôt cho d·ªÖ hi·ªÉu
    df_macro = df_macro.rename(columns={
        'NY.GDP.MKTP.KD.ZG': 'Tang_truong_GDP (%)', 
        'FP.CPI.TOTL.ZG': 'Lam_phat_CPI (%)',
        'FR.INR.LEND': 'Lai_suat_cho_vay (%)'
    })
    print(df_macro)
    print("-" * 30)
except Exception as e:
    print(f"L·ªói World Bank: {e}")

# =======================================================
# PH·∫¶N 2: D·ªÆ LI·ªÜU T·ª∂ GI√Å & CH·ª®NG KHO√ÅN - Ngu·ªìn: Vnstock
# =======================================================
try:
    print("\n2. ƒêang t·∫£i d·ªØ li·ªáu T·ª∑ gi√° USD/VND...")
    # L·∫•y t·ª∑ gi√° t·ª´ ƒë·∫ßu nƒÉm 2024
    df_fx = stock_historical_data(symbol='USD/VND', 
                                start_date='2024-01-01', 
                                end_date='2024-11-27', 
                                resolution='1D', 
                                type='index') # V·ªõi t·ª∑ gi√°, type l√† index
    
    if df_fx is not None:
        print(df_fx.tail())
    else:
        print("Kh√¥ng c√≥ d·ªØ li·ªáu t·ª∑ gi√°.")

    print("\n3. ƒêang t·∫£i d·ªØ li·ªáu C·ªï phi·∫øu VCB...")
    df_stock = stock_historical_data(symbol='VCB', 
                                start_date='2024-01-01', 
                                end_date='2024-11-27', 
                                resolution='1D', 
                                type='stock')
    print(df_stock.tail())

except Exception as e:
    print(f"L·ªói Vnstock: {e}")

print("\n--- HO√ÄN T·∫§T ---")
```
